```python
# Load libraries
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/train_data.csv')
df.head()
```

```python
plt.pie(df['PotentialFraud'].value_counts())
plt.show()
```

```python
fraud_counts = df['PotentialFraud'].value_counts()
plt.figure(figsize=(8, 4))
plt.pie(
    fraud_counts, 
    labels= ['Fraud', 'No Fraud'], 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=['#ff9999','#66b3ff'], 
    wedgeprops={'edgecolor': 'black'}
)
plt.title('Distribution of Potential Fraud Cases', fontsize= 16, fontweight= 'bold')
plt.show()
```
