### Setup

```python
# Load libraries
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in the data
df = pd.read_csv('data/train_data.csv')
df.head()
```

### Exploratory Data Analysis

```python
# Create a pie chart of potential fraud
plt.title('Percentage of Potentially Fraudulent Observations')
plt.pie(df['PotentialFraud'].value_counts(), autopct='%1.1f%%', labels=df['PotentialFraud'].value_counts().index, startangle=90)
plt.legend()
plt.show()
```

```python
fraud_counts = df['PotentialFraud'].value_counts()
plt.figure(figsize=(8, 4))
plt.pie(
    fraud_counts, 
    labels= ['Fraud', 'No Fraud'], 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=['#ff9999','#66b3ff'], 
    wedgeprops={'edgecolor': 'black'}
)
plt.title('Distribution of Potential Fraud Cases', fontsize= 16, fontweight= 'bold')
plt.show()
```
