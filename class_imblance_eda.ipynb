```python
# Load libraries
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/train_data.csv')
df.head()
```

```python
plt.pie(df['PotentialFraud'].value_counts())
plt.show()
```

```python
# Count occurrences of each category
fraud_counts = df['PotentialFraud'].value_counts()

# Plot the pie chart
plt.figure(figsize=(6, 6))
plt.pie(fraud_counts, labels= ['Fraud', 'No Fraud'], autopct='%1.1f%%', startangle=90, 
        colors=['#ff9999','#66b3ff'], wedgeprops={'edgecolor': 'black'})

# Add a title
plt.title('Distribution of Potential Fraud Cases')

# Show the plot
plt.show()
```
